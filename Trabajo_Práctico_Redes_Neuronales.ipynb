```python
import pandas as pd
data: pd.DataFrame = pd.read_csv('avanzados - Hoja 1.csv')
data
```

```python
training_inputs = data.drop(['arte', 'base de datos', 'IOT', 'ranzo', 'IA', 'videojuegos'], axis=1)
training_inputs
```

```python
training_outputs = data[['arte', 'base de datos', 'IOT', 'ranzo', 'IA', 'videojuegos']]
training_outputs
```

```python
import numpy as np
n = training_inputs.shape[1]
synaptic_weights = 2 + np.random.random((n,6)) - 1

display(synaptic_weights.shape)
synaptic_weights
```

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x) )

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
eval_NN = lambda x, w: sigmoid(np.dot(x, w))
```

```python
# 50000 epochs
for iteration in range(50000):

    # Arranca con los valores de training
    input_layer = training_inputs

    # Genera un output haciendo la suma de los valores de input multiplicados por su peso sinaptico, y aplicandole la función sigmoide para normalizar a -1, 1
    outputs = eval_NN(input_layer, synaptic_weights)

    # Calculo el error
    error = training_outputs - outputs

    # Ajusto los pesos en funcion al error y al resultado obtenido
    adjustments = error * sigmoid_derivative(outputs)
    synaptic_weights += np.dot(input_layer.T, adjustments)
```

```python
print('Synaptic weights after training')
print(synaptic_weights)

print('Outputs after training: ')
print(outputs)
```
